```python
def sort_cols(df):
    """A function sort dataframe by columns.
    
    Args:
        df(DataFrame):
    
    Returns:
        DataFrame
    """
    cols = df.columns
    cols = sort(df.cols)
    return df[cols]

def load_data(model_dir, year):
   
    frames = {}
    fil = model_dir + "pandas-frames." + str(year) + ".pickle.bz2"
    #test = pickle.load(bz2.open(fil, 'rb'))
    
    frames.update(pickle.load( bz2.open( fil, "rb" ) ))
    #test = sort_cols(test)
    #return test 
    return frames
```

### Problem 2

```python
def wins(x, a, b):
    """A function to winsorized numpy array.
    
    Args:
        x(np.array):
        a(float): lower bound
        b(float): upper bound
    
    Returns:
        np.array
    """
    for i in range(len(x)):
        if x[i] < a:
            x[i] = a
        elif x[i] > b:
            x[i] = b
    return x
```

### Problem 3

```python
def clean_nas(df):
    """A function to replace nan in numeric columns.
    
    Args:
        df(DataFrame):
    
    Returns:
        None
    """
    for col in df.columns:
        if df[col].dtype != 'object':
            df[col] = np.nan_to_num(df[col])
```
